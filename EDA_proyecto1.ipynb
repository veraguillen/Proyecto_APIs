DESARROLLO:
1.Importas las librerias necesarias y la DATA.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
EDA=pd.read_csv (r"C:\Users\veram\OneDrive\Escritorio\Data Science Henry\Proyecto individual I\ETL\EDA_DATA.csv" )
```

2.Informe general de la DATA, resumen estadistico y descriptivo.

```python
print(EDA.info())
print(EDA.describe())
print(EDA.head())
```

3.Distribucion y relaciones entre la DATA:


3.1 HISTOGRAMA DE DISTRIBUCION: Atraves del gráfico se observa que el genero "Drama" es el que tiene mayor frecuencia dentro de nuestro DATA.

```python
conteo_genero = EDA['generos'].value_counts()

plt.figure(figsize=(12, 8))
sns.barplot(x=conteo_genero.index, y=conteo_genero.values)
plt.title('Distribución de Películas por Género')
plt.xlabel('Género')
plt.ylabel('Cantidad de Películas')
plt.xticks(rotation=45)
plt.show()
```

3.2 VISUALIZACION DE LA POPULARIDAD: Podemos observar cuales son los generos de peliculas con mayor score de votos, se observa que los intereses son muy variados y sobresale el genero "accion".

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='generos', y='vote_average', data=EDA)
plt.title('Distribución de Popularidad de Películas por Género')
plt.xlabel('Género')
plt.ylabel('Popularidad')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

3.3 EVOLUCION TEMPORAL DE METRICAS: Observamos como ha variado los presupuestos he inversion para la realizacion de films, destaca la epoca de pandemia en 2020 donde se redujo consireablemnte despues de haber tenido su pico mas alto en los inicios del 2000.

```python
promedio_presupuesto_por_año = EDA.groupby('year')['presupuesto'].mean()

plt.figure(figsize=(10, 6))
promedio_presupuesto_por_año.plot()
plt.title('Evolución del Presupuesto Promedio por Año')
plt.xlabel('Año de Estreno')
plt.ylabel('Presupuesto Promedio')
plt.grid(True)
plt.show()
```

3.4 RELACION ENTRE PRESUPUESTO Y GANANCIAS: Vemos que las metricas estan proporcionalmente distribuidas en cuanto a presupuesto y ganacia, destacan un punto donde la película obtuvo perdidas considerables.

```python
plt.figure(figsize=(8, 6))
plt.scatter(EDA['presupuesto'], EDA['ganancias'], alpha=0.7, color='skyblue')
plt.title('Relación entre Presupuesto y Ganancias')
plt.xlabel('Presupuesto')
plt.ylabel('Ganancias')
plt.grid(True)
plt.show()
```

3.4 DIAGRAMA DE DISPERSION (VOTE AVERAGE vs VOTE COUNT): Podemos notas que las películas con menor puntaje son las que tienen mayor cantidad de votos, en comparacion a las mejor punteadas que tienen muy poco. Nos lleva a la conclusion que su puntaje alto se podria deberso a la falta de votaciones.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=EDA, x='vote_average', y='vote_count')
plt.title('Relación entre Score y Votos Contados')
plt.xlabel('vote_average')
plt.ylabel('Votos Contados')
plt.show()
```
